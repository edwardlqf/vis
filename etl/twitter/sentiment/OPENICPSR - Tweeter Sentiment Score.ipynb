### Openicpsr Twitter Data
* [data source](https://www.openicpsr.org/openicpsr/project/120321/version/V6/view?path=/openicpsr/120321/fcr:versions/V6/Twitter-COVID-dataset---Jan-2021)

```python
df = pd.read_csv('openicpsr/tweets_openicpsr.csv', na_values='-', infer_datetime_format=True).dropna(axis=0, how='any').reset_index(drop=True)
print(df.shape)
print(df.columns)
df.head()
```

```python
df = df[['valence_intensity', 'anger_intensity', 'fear_intensity',
       'sadness_intensity', 'joy_intensity', 'sentiment_category',
       'emotion_category', 'keyword_used', 'country_region', 'date_stamp']]
print(df.shape)
df.head()
```

```python
df['date_stamp'] = pd.to_datetime(df['date_stamp'])
```

```python
df.groupby(['country_region', 'date_stamp']).mean().to_csv('openicpsr/country_date_sentiment.csv', index=False)
```

```python
df.groupby(['date_stamp', 'country_region']).mean().to_csv('openicpsr/date_country_sentiment.csv', index=False)
```
